```python
#Basic Encapsulation
#Create a Student class with a private variable __name.
#Create getter and setter methods for __name.
#Create an object and access the name using the getter.
#Update the name using the setter.
#Verify that direct access to __name raises an error.
#--------------------------------------------------------

class Student:
    __school = "ABC School"
    def __init__(self, name, age):
        self.__name = name
        self.__age = age
    def display(self): 
        print(self.__name, self.__age, type(self).__school)
    def get_name(self): #getter method >> Only need to get one method not more than one
        return self.__name
    def update_name(self, new_name): #setter method
        self.__name = new_name
    def update_age(self, new_age):
        self.__age = new_age

    @classmethod
    def update_school(cls, new_school_name):
        cls.__school = new_school_name

std1 = Student('Nithin', 22)
std1.display()
std1.update_name('Nithin Reddy')
std1.update_age(23)
print('-'*30)
print(std1.get_name())
std1.display()
std1.update_school('XYZ')
print('-'*30)
std1.display()
print(hasattr(std1, '__name'))
#print(std1.__name) >> AttributeError: 'Student' object has no attribute '__name'
```

```python
#validaion
#Create a BankAccount class with a private __balance.
#Allow deposits only if the amount is positive.
#Allow withdrawals only if sufficient balance exists.
#Prevent the balance from becoming negative.

class BankAccount:
    def __init__(self, name, aid, balance):
        self.name = name
        self.aid = aid
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
        else:
            print(f'Amount should be positive inorder to deposit')
    def withdraw(self, amount):
        if amount <= 0:
            print('Amount should be positive')               
        elif self.__balance - amount >= 0:
            self.__balance -= amount
        else:
            print(f'Insufficient Balance') 
    def get_balance(self):
        return self.__balance

    def details(self):
        print(self.name, self.aid, self.__balance)

acc1 = BankAccount('Nithin', 1234, 10000)
acc1.details()
acc1.deposit(-10)
acc1.deposit(10000)
print(acc1.get_balance())
print('-'*30)
acc1.withdraw(-10000)
acc1.withdraw(30000)
acc1.withdraw(10000)
print(acc1.get_balance())
    
```

```python
#Multiple Private Variables:
#Create an Employee class with private variables __name and __salary.
#Create getters for both variables.
#Create a setter that rejects negative salaries.

class Employee:
    def __init__(self, name):
        self.__name = name
        self.__salary = None

    def set_salary(self, salary):
        if salary >= 0:
            self.__salary = salary
            return 'Salary updated successfully'
        else:
            return 'Salary cannot be negative'
    def get_name(self):
        return self.__name
    def get_salary(self):
        if self.__salary is not None:
            return self.__salary
        else:
            return 'Salary is not added to employee. Please add it using set_salary method'

emp1 = Employee('Nithin')
print(emp1.get_name())
print(emp1.get_salary())
print(emp1.set_salary(500000))
print(emp1.get_salary())
    
```

```python
#Create a Product class with a private price that cannot be negative.
class Product:
    def __init__(self, name):
        self.name = name
        self.__price = None

    def set_price(self, price):
        if price >= 0:
            self.__price = price
            return 'Price updated successfully'
        else:
            return 'Price cannot be negative'
    def get_price(self):
        if self.__price is not None:
            return self.__price
        else:
            return 'Price is not added to product'

p1 = Product('Tomato')
print(p1.get_price())
print(p1.set_price(-1))
print(p1.set_price(10))
print(p1.get_price())
```

Explain the difference between public, protected (single underscore), and private (double underscore) members in Python.
Public members have no underscore prefix. They can be accessed from anywhere using an object.

Protected members start with a single underscore (_). They are intended for use within the class and its subclasses. They can still be accessed from outside the class, but doing so is discouraged by convention.

Private members start with a double underscore (__). Python performs name mangling, so they cannot be accessed directly outside the class. They are inherited, but subclasses cannot directly access them using the original name because it has been mangled.

What is encapsulation?
Encapsulation is the process of bundling data (variables) and methods into a single unit (class) while controlling access to the data using public, protected, and private members.

Why do we hide data?
We hide data to protect it from unauthorized or invalid access and modification, ensuring the object's data remains consistent.

When should you use private variables?
Use private variables when data should not be accessed or modified directly from outside the class and should instead be controlled through methods.

Is Python's private access truly private?
No. Python does not provide true private access. It uses name mangling (_ClassName__variable) to discourage direct access, but the variable can still be accessed using its mangled name.
