```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

std1 = Student('Nithin', 22)
std2 = Student('Nandini', 23)
std3 = Student('Shubham', 25)

print(std1.name, std1.age)

std1.age = 23
print(std1.name, std1.age)

std1.city = 'Mulugu'
print(std1.name, std1.age, std1.city)

del std1.city
print(hasattr(std1, 'city'))

#print(std1.name, std1.age, std1.city)
#AttributeError: 'Student' object has no attribute 'city'
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

std1 = Student('Nithin', 22)
std2 = Student('Nandini', 23)
std3 = Student('Shubham', 25)

print(std1.name, std1.age)

std1.age = 23
print(std1.name, std1.age)

print(getattr(std1, 'name'))
setattr(std1, 'name', 'Nithin Reddy')
print(std1.name, std1.age)
setattr(std1, 'city', 'Mulugu')
print(std1.city)
delattr(std1, 'city')
print(hasattr(std1, 'city'))
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f'My name is {self.name}')
        print(f'My age is {self.age}')

p1 = Person('Nithin', 22)
p1.introduce()
```

```python
class Employee:
    def __init__(self, name, br):
        self.name = name
        self.branch = br
    def display(self):
        print(self.name, self.branch)

E1 = Employee('Nithin', 'Engineer')
E2 = Employee('Nandini', 'Manager')
E3 = Employee('Harshith', 'HR')

E1.display()
E2.display()
E3.display()
```

```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
    def balance_check(self):
        print(self.balance)

nithin = BankAccount('Nithin', 15000)
nithin.balance_check()
```

```python
class Rectangle:
    def area(width, height):
        return 0.5 * width * height
    def perimeter(width, height):
        return 0.5*(width + height)

res1 = Rectangle.area(5, 10)
res2 = Rectangle.perimeter(5, 10)
print(res1, res2, sep = '\n')
```

```python
class Circle:
    def circumference(radius):
        return 2*3.14*radius

print(Circle.circumference(4))
print(Circle.circumference(10))
print(Circle.circumference(3.14))
```

```python
class Student:
    def study():
        print('Student is studying')
    def sleep():
        print('Student is sleeping')

Student.study()
Student.sleep()
```

What is an attribute?
Attributes are variables (data members) that store the state or data of an object.
What is a method?
Methods are functions defined inside a class that perform operations on an object's data (attributes).
Why is self needed?
self refers to the current object. It allows each object to access and modify its own attributes and methods.
