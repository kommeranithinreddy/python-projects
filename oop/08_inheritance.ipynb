# Inheritance

This notebook contains practice problems on inheritance in Python.

## Topics Covered
- Single Inheritance
- Multilevel Inheritance
- Method Overriding
- Using `super()`
- Reusing parent class functionality

The exercises demonstrate how inheritance promotes code reuse and extensibility.

```python
#Create a Person class and a Student class that inherits from it.
#Access the parent class attributes from the child object.

class Person:
    name = None
    age = None

class Student(Person):
    course = None

std1 = Student()
std1.name = 'Nithin'
std1.age = 23
print(std1.name, std1.age)
```

```python
#Create a Vehicle class and a Car class that inherits from it.
#Create an Animal class and a Dog class that inherits from it.
#Access the parent class attributes from the child object.
class Vehicle:
    wheels = None
    steering = 1
    headlights = 2

class Car(Vehicle):
    wheels = 4

car1 = Car()
print(car1.wheels, car1.steering, car1.headlights)

class Animal:
    legs = 4
    mouth = 1
    ears = 2
    sound = None

class Dog(Animal):
    sound = 'Bark'

dog1 = Dog()
print(dog1.legs, dog1.mouth, dog1.ears, dog1.sound)
```

```python
#Access the parent class methods from the child object.
class Person:
    def talk(self):
        return 'Person is talking'

class Student(Person):
    def read(self):
        return 'Student is reading'

std1 = Student()
print(std1.read())
print(std1.talk())
```

```python
#Constructors & super()
#Create a parent class with a constructor.
#Create a child class with its own constructor.
#Use super() to call the parent constructor.
#Print values from both parent and child constructors.
#Add one extra attribute in the child class.

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Student(Person):
    def __init__(self, name, age, course):
        self.course = course
        super().__init__(name, age)

std1 = Student('Nithin', 23, 'Python')
print(std1.name, std1.age, std1.course)
```

```python
#Method Overriding
#Create a parent method display().
#Override display() in the child class.
#Call the overridden method.
#Use super().display() inside the child method.
#Observe the execution order.

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def display(self):
        return f'name = {self.name}, age = {self.age}'

class Employee(Person):
    def __init__(self, name, age, salary):
        self.salary = salary
        super().__init__(name, age)
    def display(self):        
        return f'{super().display()}, salary = {self.salary}'

emp1 = Employee('Nithin', 23, 100000)
print(emp1.display())
```

```python
#Types of Inheritance
#Single inheritance example.
class Animal:
    def sound(self):
        print(f'Animal is sounding')

class Dog(Animal):
    def sound(self):
        print(f'Dog is barking')
        super().sound()
puppy = Dog()
puppy.sound()
print('-'*30)
#Multilevel inheritance example.
class Grandfather:
    def display(self):
        print('This is Grand Father Class')

class Father(Grandfather):
    def display(self):
        print('This is Father Class')
        super().display()

class Son(Father):
    def display(self):
        print('This is Son Class')
        super().display()

son1 = Son()
son1.display()
print('-'*30)
#Hierarchical inheritance example.
class Mother:
    def display(self):
        print('This is Mother Class')

class Son(Mother):
    def display(self):
        print('This is Son Class')
        super().display()
        

class Daughter(Mother):
    def display(self):
        print('This is Daugther Class')
        super().display()

son1 = Son()
dau1 = Daughter()
son1.display()
dau1.display()
print('-'*30)
#Multiple inheritance example.
class Father:
    def display(self):
        print('This is Father Class')

class Mother:
    def display(self):
        print('This is Mother Class')

class Son(Father, Mother):
    def display(self):
            print('This is Son Class')
            super().display() # Calls Father.display()
            Mother.display(self) # Explicit call

son1 = Son()
son1.display()
print('-'*30)
#Hybrid inheritance (simple demonstration).
class Animal:
    def display(self):
        print('This is Animal Class')

class Bird(Animal):
    def display(self):
        print('This is Bird Class')
        super().display()

class Mammal(Animal):
    def display(self):
        print('This is Mammal Class')
        super().display()

class Bat(Bird, Mammal):
    def display(self):
        print('This is Bat Class')
        super().display()

bat1 = Bat()
bat1.display()
    
```

```python
#Check if a child object is an instance of the parent.
#Check if a parent object is an instance of the child.

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Student(Person):
    def __init__(self, name, age, course):
        super().__init__(name, age)
        self.course = course
    def display(self):
        print(self.course)

std1 = Student('Nithin', 23, 'Python')
per1 = Person('Nandini', 23)
print(isinstance(std1, Person)) #Because Student inherits from Person. A Student object is also a Person.
print(isinstance(std1, Student))
print('-'*30)
print(isinstance(per1, Person))
print(isinstance(per1, Student))
#Use issubclass() on different class pairs.
print('-'*30)
print(issubclass(Student, Person))
print(issubclass(Person, Student))
#Explain why the results differ.
#Student inherits from Person, so it is a subclass of Person. However, Person does not inherit from Student, so it is not a subclass of Student. 
#Inheritance is one-way.
```

```python
#Create Employee → Manager.
class Employee:
    def __init__(self, name, eid):
        self.name = name
        self.eid = eid

class Manager(Employee):
    def __init__(self, name, eid):
        super().__init__(name, eid)
        self.role = 'Manager'
        
#Create Shape → Rectangle.
class Shape:
    pass

class Rectangle(Shape):
    def __init__(self, height, lenght):
        self.height = height
        self.length = lenght


#Create Appliance → WashingMachine.
class Appliance:
    pass

class WashingMachine(Appliance):
    def __init__(self, load_capacity):
        self.load_capacity = load_capacity
        
#Create Bird → Eagle.
class Bird:
    pass
class Eagle(Bird):
    pass

#Create Book → Ebook.
class Book:
    def details(self):
        print('Book have pages')

class Ebook(Book):
    def details(self):
        print('Ebook available in online')
```
