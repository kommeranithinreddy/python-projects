```python
#Create a Student class and print an object without defining __str__(). Observe the output.
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

std1 = Student('nithin', 23)
print(std1)
```

```python
#Implement __str__() to display the student's name and age.
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __str__(self):
        return f'name : {self.name}, age = {self.age}'

std1 = Student('nithin', 23)
print(std1)
```

```python
#Implement __repr__() and compare its output with __str__().

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __repr__(self):
        return f'Student(name : {self.name}, age = {self.age})'

std1 = Student('nithin', 23)
print(repr(std1))
```

Explain the difference between __str__() and __repr__().

```python
#Create a Book class and overload the + operator using __add__().

class Book:
    def __init__(self, name, page_count):
        self.name = name
        self.page_count=page_count
    def __add__(self, other):
        return self.page_count + other.page_count

book1 = Book('Atomic Habits', 233)
book2 = Book('IKIGAi', 256)

print(book1+book2)
```

```python
#Create a BankAccount class where adding two accounts returns the total balance.
class BankAccount:
    def __init__(self, accno, balance):
        self.accno= accno
        self.balance = balance
    def __add__(self, other):
        if isinstance(other, BankAccount):
            return self.balance + other.balance
        else:
            return 'NotImplemented'

acc1 = BankAccount(1234, 20000)
acc2 = BankAccount(9876, 15000)

print(acc1 + acc2)
```

```python
#Create a Vector class and overload +.

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __str__(self):
        return f'({self.x}, {self.y})'
    def __add__(self, other):
        if isinstance(other, Vector):
            x = self.x + other.x
            y = self.y + other.y
            return Vector(x,y)
v1 = Vector(1,2)
v2 = Vector(4,5)
v3 = v1+v2
print(v1, v2, v3)
```

```python
#Create a Rectangle class and overload == using __eq__().

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def __eq__(self, other):
        if isinstance(other, Rectangle):
            if self.width == other.width and self.height == other.height:
                return True
            else:
                return False
        else:
            return False

rect1 = Rectangle(5, 10)
rect2 = Rectangle(10, 15)
rect3 = Rectangle(5, 10)

print(rect1 == rect2)
print(rect2 == rect3)
print(rect3 == rect1)
```

```python
#Create a Person class and overload < using __lt__() based on age.
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __lt__(self, other):
        if isinstance(other, Person):
            if self.age<other.age:
                return True
            else:
                return False
        else:
            return 'NotImplemented'

p1 = Person('Nithin', 23)
p2 = Person('Nandini', 24)
p3 = Person('Shubham', 25)

print(p1<p2)
print(p2<p3)
print(p3<p1)
```

When implementing magic methods, prefer returning NotImplemented over returning custom strings for unsupported types. \
It's the Pythonic approach and allows Python to handle unsupported operations properly.

```python
#Create a Playlist class and implement __len__().
class Playlist:
    def __init__(self, *args):
        self.songs = list(args)
    def __len__(self):
        count = 0
        for i in self.songs:
            count += 1
        return count
p1 = Playlist('s1', 's2', 's3', 's4', 's5', 's6', 's7', 's8', 's9', 's10', 's11', 's12')
print(len(p1))
```

```python
#Create a Cart class and implement __len__() to return the number of items.
class Cart:
    def __init__(self, *args):
        self.items = args
    def __len__(self):
        count = 0
        for i in self.items:
            count += 1
        return count
c1 = Cart('Laptop', 'Mobile', 'Tab', 'charger', 'Mouse', 'Keyboard', 'headset')
print(len(c1))
```

```python
#Create a Wallet class and implement __bool__() so it returns False when the balance is zero.

class Wallet:
    def __init__(self, balance):
        self.balance = balance
    def __bool__(self):
        return bool(self.balance)

w1 = Wallet(10)
w2 = Wallet(0)
w3 = Wallet(1000)
wallets = [w1, w2, w3]

for wallet in wallets:
    if wallet:
        print('True')
    else:
        print('False')
```

```python
#Create a Team class and make it evaluate to True only if it has at least one player.

class Team:
    def __init__(self, *args):
        self.players = args
    def __bool__(self):
        return len(self.players) != 0

t1 = Team('Nithin', 'Nandini', 'Shubham', 'Santhosh', 'Nagaraju', 'Anil')
t2 = Team()
t3 = Team('Harshith', 'Vignesh', 'Sudhamsh')

teams = [t1, t2, t3]
for team in teams:
    if team:
        print('True')
    else:
        print('False')
```

Explain when Python calls __bool__().
Python calls __bool__() whenever it needs to determine the truth value of an object, such as in conditional statements (if, while) or when the bool() function is used. 
If __bool__() is not defined, Python falls back to __len__().

```python
#Create a class implementing __call__() so an object can be called like a function.

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __call__(self):
        print(f'{self.name} is {self.age} years old')

p1 = Person('Nithin', 23)
p1()
```

```python
#Create a Temperature class implementing __int__() and __float__().

class Temperature:
    def __init__(self, temp):
        self.temp = temp
    def __int__(self):
        return int(self.temp)
    def __float__(self):
        return float(self.temp)

t1 = Temperature(91.8)
print(int(t1))
print(float(t1))
```

```python
#Create a class implementing __contains__() so you can use the in operator.
class Team:
    def __init__(self, *args):
        self.players = args
    def __contains__(self, player):
        return player in self.players

T1 = Team('Nithin', 'Nandini', 'Shubham')
print('Nithin' in T1)
print('Santhosh' in T1)
```

What are magic methods?
Magic methods (dunder methods) are special methods with double underscores that Python automatically calls in specific situations, such as object creation, operator usage, or built-in function calls.

Why are they called dunder methods?
They are called dunder methods because their names begin and end with double underscores (__), where "dunder" means "double underscore."

What is operator overloading?
Operator overloading is the process of giving operators like +, ==, and < custom behavior for user-defined objects by implementing the corresponding magic methods.
