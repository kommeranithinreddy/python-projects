# Polymorphism

This notebook contains practice problems demonstrating polymorphism in Python.

## Topics Covered
- Method overriding
- Runtime polymorphism
- Duck typing
- Common interfaces

The exercises illustrate how different objects can respond differently to the same method call.

```python
#Create an Animal class with sound().
class Animal:
    def sound(self):
        print('Animal is sounding')
#Create Dog and override sound().
class Dog(Animal):
    def sound(self):
        print('Dog is sounding: Bow Bow')
#Create Cat and override sound().
class Cat(Animal):
    def sound(self):
        print('Cat is sounding: Meow Meow')
#Create Cow and override sound().
class Cow(Animal):
    def sound(self):
        print('Cow is sounding: Moo Moo')
#Call sound() using objects of all three classes.
dog1 = Dog()
cat1 = Cat()
cow1 = Cow()
animals = [dog1, cat1, cow1]
for animal in animals:
    animal.sound()
```

```python
#Create Shape with area().
class Shape:
    def area(self):
        pass
#Create Rectangle and implement area().
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
#Create Circle and implement area().
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return self.radius * self.radius * 3.14

#Store both objects in a list and call area() using a loop.
#Observe how the same method behaves differently.
rect1 = Rectangle(5, 10)
circ1 = Circle(8)
shapes = [rect1, circ1]
for shape in shapes:
    print(shape.area())
```

```python
#Create a Bird class with fly().
class Bird:
    def fly(self):
        print('Bird is flying')

#Create an Airplane class with fly().
class Airplane:
    def fly(self):
        print('Airplane is flying')

#Write a function that accepts any object and calls fly().
def start_fly(obj):
    obj.fly()

bird1 = Bird()
airp1 = Airplane()
objs = [bird1, airp1]
#Pass both Bird and Airplane objects to the function.
for obj in objs:
    start_fly(obj)
#Add another class (Drone) with fly() and reuse the same function.
class Drone:
    def fly(self):
        print('Drone is flying')
drone1 = Drone()
start_fly(drone1)
```

```python
#Show how len() works with a string, list, tuple, and dictionary.
str1 = 'Nithin'
lst1 = [1,2,3,4,5,6]
tup1 = (1,2,3,4,5,6,7)
dict1 = {'nithin':1, 'nandini':2, 'shubham':3, 'santhosh':4}
print(len(str1))
print(len(lst1))
print(len(tup1))
print(len(dict1))

#For a string, it counts characters.
#For a list or tuple, it counts elements.
#For a dictionary, it counts keys.
```

```python
#Show how the + operator behaves differently for integers, strings, and lists.
int1 = 10
int2 = 20
print(int1 + int2)

str1 = "Nithin"
str2 = " Reddy"
print(str1 + str2)

list1 = [1, 2]
list2 = [3, 4]
print(list1 + list2)

#int → Adds numbers.
#str → Joins strings.
#list → Combines lists.
```
