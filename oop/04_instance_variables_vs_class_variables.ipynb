```python
#Instance Variables
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def display(self):
        print(self.name, self.age)

std1 = Student('Nithin', 22)
std2 = Student('Shubham', 25)
std3 = Student('Santhosh', 24)

std1.display()
std2.display()
std3.display()

std2.age += 1
print('-'*30)
std1.display()
std2.display()
std3.display()

std1.city = 'Mulugu'
print(std1.city)

del std3.name
print('-'*30)
std1.display()
std2.display()
#std3.display() >> AttributeError: 'Student' object has no attribute 'name'
print(hasattr(std3, 'name'))
```

```python
#class Variables

class Student:
    school = 'ABC School'
    student_count = 0
    def __init__(self):
        Student.student_count += 1

std1 = Student()
std2 = Student()
std3 = Student()

print(std1.school)
print(std2.school)
print(std3.school)

Student.school = 'XYZ School'
print('-'*30)
print(std1.school)
print(std2.school)
print(std3.school)

Student.country = 'India'
print('-'*30)
print(std1.country)
print(std2.country)
print(std3.country)

print(Student.student_count)
```

```python
#Comparing Instance & Class Variables

class Car:
    wheels = 4
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price
    def display(self):
        print(self.__dict__)
        print(f'Wheels = {self.wheels}')

car1 = Car('benz', 2500000)
car2 = Car('Nano', 100000)
car3 = Car('Maruti', 250000)

car1.display()
car2.display()
car3.display()

car2.brand = 'Alto'
print('-'*30)

car1.display()
car2.display()
car3.display()

Car.wheels = 6
print('-'*30)

car1.display()
car2.display()
car3.display()
print('-'*30)

car1.wheels = 4
car1.display()
car2.display()
car3.display()
print('-'*30)

del car1.wheels

car1.display()
car2.display()
car3.display()
```

```python
#practice

class Employee:
    company = 'Google'
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

class BankAccount:
    bank_name = 'Axis'
    def __init__(self, acc_no, name, balance):
        self.acc_no = acc_no
        self.name = name
        self.balance = balance

class Book:
    library_name = 'Bits-Library'
    def __init__(self, name, author, price):
        self.name = name
        self.author = author
        self.price = price
e1 = Employee('Nithin', 36000000)
e2 = Employee('Shubham', 35000000)

acc1 = BankAccount(1234, 'nithin', 100000)
acc2 = BankAccount(5678, 'nandini', 2500000)

book1 = Book('Atomic Habits', 'James Clear', 600)
book2 = Book('The Monk', 'Robin Sharma', 250)


print('Class : ', Employee.__dict__.keys())
print('Object : ', e1.__dict__)

print('Class : ',BankAccount.__dict__.keys())
print('Object : ', acc1.__dict__)

print('Class : ',Book.__dict__.keys())
print('Object : ', book1.__dict__)
```

What is an instance variable?
Instance variables are object-level variables. They are created for each object (usually inside __init__()) and each object has its own copy. They are typically accessed through an object.

What is a class variable?
Class variables belong to the class and are shared by all objects. They can be accessed using the class name or an object.

When should each be used?
Use class variables for data shared by all objects. Use instance variables for data that is unique to each object.

Why does changing an instance variable not affect other objects?
Because each object has its own separate copy of instance variables. Changing one object's instance variable does not affect the instance variables of other objects.

Why does changing a class variable affect all objects?
It's class level variable, updating with class name only affects in all objects.
