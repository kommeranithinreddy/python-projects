```python
#instance methods:

class Student:
    def __init__(self, name, age, course):
        self.name = name
        self.age = age
        self.course = course
    def display_details(self):
        print(self.name, self.age)


std1 = Student('Nithin', 22, 'Data Science')
std2 = Student('Shubham', 25, 'Ai Scientist')
std3 = Student('Harshiht', 23, 'HR')

std1.display_details()
std2.display_details()
std3.display_details()
```

```python
class Employee:
    def __init__(self, name, eid, salary):
        self.name = name
        self.eid = eid
        self.salary = salary
    def details(self):
        print(self.name, self.eid, self.salary)
    def increase_salary(self, percentage):
        self.salary += self.salary*percentage//100

emp1 = Employee('Nithin', 52148783, 25000)
emp2 = Employee('Shubham', 52148784, 25000)

emp1.details()
emp2.details()
print('-'*40)
emp1.increase_salary(200)
emp1.details()
emp2.details()
```

```python
class BankAccount:
    def __init__(self, accno, name, balance):
        self.accno = accno
        self.name = name
        self.balance = balance

    def details(self):
        print(self.accno, self.name, self.balance)

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if self.balance > amount:
            self.balance -= amount
        else:
            print('Insufficient Balance')

acc1 = BankAccount(1234, 'Nithin', 50000)
acc2 = BankAccount(9987, 'Shubham', 30000)

acc1.details()
acc2.details()
print('-'*30)
acc1.deposit(30000)
acc2.deposit(20000)
acc1.details()
acc2.details()
print('-'*30)
acc1.withdraw(20000)
acc2.withdraw(1000000)
acc1.details()
acc2.details()
```

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def perimeter(self):
        return 2*self.width* self.height

    def area(self):
        return self.width * self.height

rect1 = Rectangle(5, 10)
rect2 = Rectangle(2, 4)

res1 = rect1.perimeter()
res2 = rect1.area()
res3 = rect2.perimeter()
res4 = rect2.area()
print(f'perimeter of rect1 = {res1}')
print(f'area of rect1 = {res2}')
print(f'perimeter of rect2 = {res3}')
print(f'area of rect2 = {res4}')
```

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14 * self.radius*self.radius
    def circumference(self):
        return 2*3.14*self.radius

circle1 = Circle(7)
res1 = circle1.area()
res2 = circle1.circumference()

print(f'area of circle1 = {res1}')
print(f'area of circle2 = {res2}')
```

```python
class Student:
    school = 'ABC School'
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def change_school(cls, new_name):
        cls.school = new_name

    def details(self):
        print(self.name, self.age, self.school)

std1 = Student('Nithin', 23)
std2 = Student('Nandini', 23)

std1.details()
std2.details()
print('-'*40)
std1.change_school('XYZ School')
std1.details()
std2.details()
```

```python
#Create a class variable student_count and update it whenever an object is created.
#Create a class method that returns the total number of objects created.
#Verify that a class method can be called using both the class and an object.

class Student:
    school = 'ABC'
    student_count = 0
    def __init__(self, name, age):
        self.name = name
        self.age = age
        Student.student_count += 1 #type(self).student_count += 1
    
    def details(self):
        print(self.name, self.age, self.school)

    @classmethod
    def total_student_count(cls):
        print(cls.student_count)


std1 = Student('Nithin', 23)
std2 = Student('Shubham', 25)
std3 = Student('Nandini', 23)
std4 = Student('Santhosh', 24)

std1.details()
std2.details()

Student.total_student_count() #called by class
std1.total_student_count() #called by object
```

```python
#Static Methods (@staticmethod)
#Create a static method is_even(number).
#Create a static method is_prime(number).
#Create a static method to calculate the square of a number.

class Number:
    @staticmethod
    def is_even(num):
        if num%2 == 0:
            return True
        return False

    @staticmethod
    def is_prime(num):
        count = 0
        for i in range(1, num+1):
            if num%i == 0:
                count += 1

        if count == 2:
            return True
        return False

    @staticmethod
    def square(num):
        return num**2

res1 = Number.is_even(4)
res2 = Number.is_prime(11)
res3 = Number.square(9)

print(res1, res2, res3, sep = '\n')
```

```python
#Create a static method to convert Celsius to Fahrenheit.
class Temperature:
    @staticmethod
    def to_fahrenheit(celsius):
        return (celsius * 9/5) + 32

    @staticmethod
    def to_celsius(fahrenheit):
        return (fahrenheit - 32) * 5/9

res1 = Temperature.to_fahrenheit(32)
res2 = Temperature.to_celsius(98)
print(res1, res2)
```

```python
#Create a static method that checks whether a given year is a leap year.

class Year:
    @staticmethod
    def is_leap(year):
        if year%400 == 0:
            return True
        elif year%4 == 0 and year%100 != 0:
            return True
        return False
            
print(Year.is_leap(2024))
print(Year.is_leap(2000))
print(Year.is_leap(2100))
print(Year.is_leap(1600))
```

```python
#Create a Student class that contains:
#Instance method → display student details
#Class method → change school name
#Static method → validate age (between 5 and 100)


class Student:
    school = 'ABC'
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(self.name, self.age, self.school)
        
    @classmethod
    def change_school(cls, new_name):
        cls.school = new_name

    @staticmethod
    def validate_age(age):
        return 5 <= age <= 100
             

std1 = Student('Nithin', 23)
std1.display()
std1.change_school('XYZ')
print('-'*30)
std1.display()
print(Student.validate_age(2))
print(std1.validate_age(10))
```

Explain why self, cls, and no first parameter are used in instance, class, and static methods respectively.
self refers to the current object.
cls refers to the current class.
Static methods don't need either because they don't operate on object or class data.

Instance method
Use an instance method when the operation depends on a particular object's data. It can access both instance variables and class variables. It is typically called using an object.

Class method
Use a class method when the operation is related to the class as a whole rather than a specific object. It primarily accesses and modifies class variables. It can be called using either the class or an object.

Static method
Use a static method for utility functions that don't depend on object or class state. It doesn't automatically receive self or cls, so it cannot directly access instance data and only accesses class data if you explicitly reference the class name.
