```python
#Create an Engine class and a Car class. A Car should create its own Engine.
class Engine:
    def start(self):
        print('Engine started .....')
    def stop(self):
        print('Engine stopped .....')

class Car:
    def __init__(self):
        self.engine = Engine()
    def start(self):
        self.engine.start()
    def stop(self):
        self.engine.stop()

car1 = Car()
#Add a method in Car that calls a method from Engine.
car1.start()
car1.stop()
```

```python
#Create a CPU class and a Computer class where the Computer owns the CPU.
class CPU:
    pass

class Computer:
    def __init__(self):
        self.cpu = CPU()
comp1 = Computer()
```

```python
#Create a Heart class and a Human class where the Human owns the Heart.

class Heart:
    pass

class Human:
    def __init__(self):
        self.heart = Heart()

h1 = Human()
```

```python
#Create a Teacher class and a Department class. Pass a Teacher object into the Department.

class Teacher:
    pass

class Department:
    def __init__(self, obj):
        self.teacher = obj

t1 = Teacher()
dp1 = Department(t1)
```

```python
#Create a Player class and a Team class. A Team should receive Player objects.

class Player:
    pass

class Team:
    def __init__(self, team_name, *args):
        self.team_name = team_name
        self.players_list = list(args)

p1 = Player()
p2 = Player()
p3 = Player()
p4 = Player()
p5 = Player()

t1 = Team('Riders', p1,p2,p3,p4,p5)
```

Answer these questions:
What is composition?
Composition is a strong has-a relationship where one object creates and owns another object.
What is aggregation?
Aggregation is a weak has-a relationship where one object is created independently and then passed to another object.
What is a "has-a" relationship?
Both composition and aggregation are has-a relationships.

Examples:

Car has an Engine → Composition
Department has Teachers → Aggregation

A has-a relationship means one class contains or owns another class as part of its design. Both composition and aggregation are has-a relationships.

What is the main difference between composition and aggregation?
Composition owns the contained object, whereas aggregation only references an independently existing object.
