```python
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price

class Employee:
    def __init__(self, name, id, salary):
        self.name = name
        self.id = id #id only exists inside the function. It doesn't break Python's built-in id() function globally.
        self.salary = salary

class Laptop:
    def __init__(self, brand, ram, storage):
        self.brand = brand
        self.ram = ram
        self.storage = storage

class Mobile:
    def __init__(self, brand, model, battery):
        self.brand = brand
        self.model = model
        self.battery = battery

class BankAccount:
    def __init__(self, acc_num, balance):
        self.account_number = acc_num
        self.balance = balance

book1 = Book('Atomic Habits', 'James Clear', 600)
emp1 = Employee('Nithin', 52148783, 23000)
laptop1 = Laptop('Lenovo', 24, 1024)
mobile1 = Mobile('Oneplus', 'Nord CE3', '5000mAh')
acc1 = BankAccount(12345, 30000)

print(book1.title, book1.author, book1.price)
print(emp1.id, emp1.name, emp1.salary)
print(laptop1.brand, laptop1.ram, laptop1.storage)
print(mobile1.brand, mobile1.model, mobile1.battery)
print(acc1.account_number, acc1.balance)
```

```python
#Create two students and swap their ages.

class Student:
    def __init__(self, name, age, course):
        self.name = name
        self.age = age
        self.course = course

    def display(self):
        print(self.name, self.age, self.course)

std1 = Student('Nithin Reddy', 22, 'Python')
std2 = Student('Nandini J',  23, 'Quantum Physics')
std1.display()
std2.display()

std1.age, std2.age = std2.age, std1.age
print('-' * 20)
std1.display()
std2.display()
```

```python
#Create three employees and increase each salary by 10%

class Employee:
    def __init__(self, name, id, salary):
        self.name = name
        self.id = id 
        self.salary = salary

    def display(self):
        print(self.name, self.id, self.salary)

emp1 = Employee('Nithin', '52148783', 23000)
emp2 = Employee('Shubham', '52148784', 22000)
emp3 = Employee('Santhosh', '52148785', 24000)

emp1.display()
emp2.display()
emp3.display()

employees = [emp1, emp2, emp3]

for emp in employees:
    emp.salary += emp.salary*10/100


emp1.display()
emp2.display()
emp3.display()
```

```python
#Create a bank account object and deposit ₹5000.
#Create a bank account object and withdraw ₹2000.

class BankAccount:
    def __init__(self, name, id, balance):
        self.name = name
        self.id = id
        self.balance = balance
    def details(self):
        print(self.name, self.id, self.balance)

acc1 = BankAccount('Nithin', 1234, 50000)

acc1.details()
acc1.balance += 20000
print('-' * 30)
acc1.details()
acc1.balance -= 40000
print('-' * 30)
acc1.details()
```

```python
#Create a Rectangle class and calculate its area.

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

rc1 = Rectangle(10, 20)
print(rc1.area())
```

```python
#Create ten objects using a loop and store them in a list.
#Print every object's attributes using a loop.

class Chr:
    def __init__(self, num, char):
        self.num = num
        self.char = char

objects = []
for i in range(65, 75):
    obj = Chr(i, chr(i))
    objects.append(obj)

for obj in objects:
    print(obj.num, obj.char)
```

What is a constructor?
A constructor is the __init__() method. It is automatically called when an object is created and is used to initialize the object's instance variables.

When is __init__() called?
__init__() is called automatically when an object is created.

Can one class have multiple objects with different constructor values?
Yes. Each object can be created with different constructor arguments, so every object can have its own attribute values.

Why do we use constructors instead of assigning attributes manually?
Constructors initialize an object at the time it is created. This makes object creation simpler, ensures required data is provided, and avoids assigning attributes manually after creating the object.
