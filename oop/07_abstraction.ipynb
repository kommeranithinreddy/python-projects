This notebook contains practice problems on abstraction using Python's `abc` module.

Each exercise demonstrates how abstraction enforces a common interface while allowing different implementations.

```python
#Create an abstract class Animal with an abstract method sound().
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        print('Animal is sounding')

#Create a Dog class that implements sound().
class Dog(Animal):
    def sound(self):
        print('Bow Bow')

#Create a Cat class that implements sound().
class Cat(Animal):
    def sound(self):
        print('Meow meow')

#Create objects of Dog and Cat and call sound().
dog1 = Dog()
cat1 = Cat()
dog1.sound()
cat1.sound()

#Try creating an object of Animal. Observe and explain the error.
'''animal1 = Animal()
TypeError: Can't instantiate abstract class Animal without an implementation for abstract method 'sound'
Because Animal is an abstract class.
An abstract class is incomplete. It defines what subclasses must implement, but it is not meant to be instantiated directly.
Python prevents this by raising a TypeError.
Only classes like Dog and Cat that implement all abstract methods can be instantiated.'''
```

```python
#Create an abstract class Shape with abstract methods area() and perimeter().
from abc import ABC, abstractmethod
class Shape(ABC):
    def __init__(self):
        pass
    @abstractmethod
    def area(self):
        pass
    @abstractmethod
    def perimeter(self):
        pass

#Create a Rectangle class implementing both methods.
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    def perimeter(self):
        return 2*(self.width + self.height)

#Create a Circle class implementing both methods.
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14*self.radius*self.radius
    def perimeter(self):
        return 2*3.14*self.radius
rect1 = Rectangle(5, 10)
circ1 = Circle(8)
#Store both objects in a list and call their methods using a loop.
shapes = [rect1, circ1]

for shape in shapes:
    print(shape.area())
    print(shape.perimeter())

#Explain why every subclass must implement all abstract methods.
#Every subclass must implement all abstract methods because an abstract class defines a contract. 
#Until all abstract methods are implemented, the subclass remains incomplete, so Python does not allow objects of that subclass to be created.
```

```python
#Create an abstract class Payment with an abstract method pay(amount).
#Implement CreditCardPayment
#Implement UPIPayment

from abc import ABC, abstractmethod
class Payment(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

class CreditCardPayment(Payment):
    def pay(self, amount):
        print(f'Amount {amount} is paid via credit card')

class UPIPayment(Payment):
    def pay(self, amount):
        print(f'Amount {amount} is paid via UPI')

#Create an abstract class Vehicle with abstract methods start() and stop().
#Implement Car
#Implement Bike

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
    @abstractmethod
    def stop(self):
        pass

class Car(Vehicle):
    def start(self):
        print('Car is started')
    def stop(self):
        print('Car is stopped')

class Bike(Vehicle):
    def start(self):
        print('Bike is started')
    def stop(self):
        print('Bike is stopped')

#Create an abstract class Employee with an abstract method calculate_salary().
#Implement FullTimeEmployee
#Implement PartTimeEmployee
class Employee(ABC):
    @abstractmethod
    def calculate_salary(self):
        pass
class FullTimeEmployee(Employee):
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    def calculate_salary(self):
        return self.salary

class PartTimeEmployee(Employee):
    def __init__(self, name, hourly_rate):
        self. name = name
        self.hourly_rate = hourly_rate
    def calculate_salary(self, hours_worked):
        return hours_worked * self.hourly_rate

#Create an abstract class Notification with an abstract method send(message).
#Implement EmailNotification
#Implement SMSNotification
class Notification(ABC):
    @abstractmethod
    def send(self, message):
        pass

class EmailNotification(Notification):
    def send(self, message):
        print(f'{message} is sent in Email')

class SMSNotification(Notification):
    def send(self, message):
        print(f'{message} is sent in SMS')
        
```
