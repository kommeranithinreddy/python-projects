- Creating classes
- Creating objects

- Object methods

Each problem is solved independently to strengthen the understanding of classes and objects.

```python
class Student:
    pass

nithin = Student()
nandini = Student()
shubham = Student()

print(nithin, nandini, shubham)
print(type(nithin), type(nandini), type(shubham), sep = '\n')
print(id(nithin), id(nandini), id(shubham), sep = '\n')
```

```python
class Car:
    pass

car1 = Car()
car2 = Car()
car3 = Car()
car4 = Car()
car5 = Car()

print(car1, car2, car3, car4, car5)
```

```python
class Book:
    pass

class Laptop:
    pass

class Mobile:
    pass

book1 = Book()
book2 = Book()
laptop1 = Laptop()
laptop2 = Laptop()
mobile1 = Mobile()
mobile2 = Mobile()

print(type(book1), type(book2), type(laptop1), type(laptop2), type(mobile1), type(mobile2), sep = '\n')
```

```python
class Book:
    pass


book1 = Book()
book2 = Book()
print(book1 == book2)
print(book1 is book2)
```

```python
class Student:
    pass

nithin = Student()
print(isinstance(nithin, Student))
```

```python
class Book:
    pass

print(dir(Book))
print()
print(Book.__mro__)
```

```python
class Car:
    pass

car1 = Car()
car2 = Car()
car3 = Car()
car4 = Car()
car5 = Car()
car_list = [car1, car2, car3, car4, car5]

for car in car_list:
    print(type(car))

for car in car_list:
    print(id(car))
```

```python
class Animal:
    pass

animals_list = []
for _ in range(10):
    animal = Animal()
    animals_list.append(animal)

print(animals_list)
print(len(animals_list))
```

```python
class Car:
    pass

car1 = Car()
car2 = car1

print(car1 == car2)
print(car1 is car2)
print(id(car1))
print(id(car2))
```

```python
class Circle:
    pass

class Triangle:
    pass

class Rectangle:
    pass

circle1 = Circle()
triangle1 = Triangle()
rectangle1 = Rectangle()

print(type(circle1))
print(type(triangle1))
print(type(rectangle1))
```

```python
class Human:
    pass

class Animal:
    pass

class Reptail:
    pass

class Bird:
    pass

human1 = Human()
animal1 = Animal()
reptail1 = Reptail()
bird1 = Bird()

species = [human1, animal1, reptail1, bird1]

for unit in species:
    print(type(unit).__name__)
```

```python
class Person:
    pass

nithin = Person()

for value in dir(nithin):
    print(value)
```
